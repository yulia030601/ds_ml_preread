```python
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
import datetime

np.random.seed(123)

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (16, 8))
f.suptitle('Plot examples', fontsize = 20)


#ax1
dots = np.random.multivariate_normal([20, 10], [[1, 0.7], [0.7, 1]], size = 1000)

ax1.scatter(dots[:, 0], dots[:, 1], edgecolors = 'black')
ax1.set_title('Scatter plot example', )

#ax2
xs2 = np.random.geometric(p=0.1, size=3000)
ax2.hist(xs2, bins = 40, color = 'r')
ax2.set_title('Hist plot example', )

#ax3
xs3 = np.linspace(0, 6)
ys3_1 = np.log(xs3)
ys3_2 = np.log(2*xs3)
ax3.plot(xs3, ys3_1, color = 'black')
ax3.plot(xs3, ys3_2, color = 'red')
ax3.set_title('Line plot example', )
ax3.legend(['log (x)', 'log (2x)'])


#ax4
base = datetime.date(2021, 1, 1)
xs4 = [base + datetime.timedelta(days=x) for x in range(14)]
ys4 = np.random.normal(loc = 5, scale = 1, size = 14)
ax4.bar(xs4, height = ys4)
ax4.set_title('Bar plot example', )
ax4.set_xticks(xs4)
plt.xticks(rotation=90)
```
